```python
# -*- coding: utf-8 -*-
def get_oneurl(url):
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    from selenium.webdriver.common.keys import Keys
    from selenium.webdriver.support.ui import Select
    from selenium.common.exceptions import NoSuchElementException
    from selenium.common.exceptions import NoAlertPresentException
    import unittest, time, re
    import requests
    from bs4 import BeautifulSoup as bs

    driver = webdriver.Firefox()
    driver.implicitly_wait(30)
    driver.get(url)
    soup = bs(driver.page_source, 'xml')

    for ele in soup.select('table#Table1 tr'):
        if len(ele.select('a')) > 0:
             if re.search('\/*\/',  ele.select('a')[0]['href']) == None and re.search('\&*\%',  ele.select('a')[0]['href']) == None:
                print 'http://www.liontravel.com/webpd/' + ele.select('a')[0]['href']
                get_data('http://www.liontravel.com/webpd/' + ele.select('a')[0]['href'])
                time.sleep(1)
```

- 多個網頁

```python
# -*- coding: utf-8 -*-
def get_nexturl(url):
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    from selenium.webdriver.common.keys import Keys
    from selenium.webdriver.support.ui import Select
    from selenium.common.exceptions import NoSuchElementException
    from selenium.common.exceptions import NoAlertPresentException
    import unittest, time, re
    import requests
    from bs4 import BeautifulSoup as bs

    driver = webdriver.Firefox()
    driver.implicitly_wait(30)
    driver.get(url)
    soup = bs(driver.page_source, 'xml')

    num = int(soup.select('span a')[77]['data-pageindex'])

    for i in range (0, num):
        try:
            soup = bs(driver.page_source, 'xml')
            for ele in soup.select('table#Table1 tr'):
                if len(ele.select('a')) > 0:
                    if re.search('\/*\/',  ele.select('a')[0]['href']) == None and re.search('\&*\%',  ele.select('a')[0]['href']) == None:
                        print 'http://www.liontravel.com/webpd/' + ele.select('a')[0]['href']
                        get_data('http://www.liontravel.com/webpd/' + ele.select('a')[0]['href'])
                        time.sleep(1)
            driver.find_element_by_link_text(u"下一頁").click()
            time.sleep(3)
        except:
            break
```

- 細項資料

```python
def get_data(url):
    import requests
    from bs4 import BeautifulSoup as bs
    
    global tourlist  #tourlist設為global為全域變數
    global num   #num設為global為全域變數
    dic={"agency":"", "title":"", "prodNo1":"", "prodNo2":"", "tour":{}, "href":""}
    
    rs = requests.session()
    res = rs.get(url)
    soup = bs(res.text, "lxml")
    num += 1  #計算總共行程數

    #title
    dic["title"] = soup.select('span.gray.fsxxl.title')[0].text

    #agency
    dic["agency"] = u"雄獅旅遊"

    #prodNo  將網址切顆只剩下prodNo 其他旅行社有prodNo2,但這裡沒有,所以prodNo2要設為null
    Prod = url.strip().split('?sProd=')[1]
    dic["prodNo1"] = Prod
    dic["prodNo2"] = 'Null'

    #href
    dic["href"] = url

    #tour    每天的行程不同，用day1..day+n來區分
    for ele in soup.select('div.tblBody table.tbl_ltrc'):
        for i in range (0, len(ele.select('p'))):
            dic["tour"]['Day'+str(i+1)] = ele.select('p')[i].text

    tourlist.append(dic)  #將dic加入tourlist裡
```
