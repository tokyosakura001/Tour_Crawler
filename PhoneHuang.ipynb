```python
def get_url(url):
    import requests
    from bs4 import BeautifulSoup as bs
    import time

    rs = requests.session()  #變為無痕模式
    res = rs.get(url)
    soup =  bs(res.text, 'lxml')

    for ele in soup.select('table.trip_list tr'):
        if len(ele.select('a')) > 0:
            print ele.select('a')[0]['onclick'].split("'")[1]
            get_data(ele.select('a')[0]['onclick'].split("'")[1])
            time.sleep(1)
```

```python
def get_data(url):
    from bs4 import BeautifulSoup as bs
    import requests
    global tourlist   #tourlist設為global為全域變數

    res = requests.get(url)
    soup = bs(res.text, 'lxml')

    #title  有寫title是寫在table h1裡，有些寫在#headLeft h1，用if elif來做title在哪裡的判斷
    if len(soup.select('table h1')) > 0:  
        dic["title"] = soup.select('table h1')[0].text
    elif len(soup.select('#headLeft h1')) > 0:
        dic["title"] = soup.select('#headLeft h1')[0].text
    
    #agency
    dic["agency"] = u"鳳凰旅遊"
    
    #PID     將網址切顆只剩下GROUP_NO, TOUR_NO
    rearstring = url.strip().split('GROUP_NO=')  
    GROUP_NO, TOUR_NO =  rearstring[1].split('&TOUR_NO=')
    dic["prodNo1"] = GROUP_NO
    dic["prodNo2"] = TOUR_NO    
        
    #href   
    dic["href"] = url
        
    #tour     每天的行程不同，用day1..day5來區分
    for ele in soup.select('#repeatBg'):
        for i in range(0, len(ele.select('.daily_text'))):
            dic["tour"]['Day'+str(i+1)] = ele.select('.daily_text')[i].text.split()[3]

    tourlist.append(dic)    #將dic加入tourlist裡
```
