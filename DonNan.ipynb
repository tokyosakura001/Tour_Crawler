- 細項資料

```python
def get_data(url):
    import requests
    from bs4 import BeautifulSoup as bs
    
    global tourlist  #tourlist設為global為全域變數
    global num  #num設為global為全域變數
    dic={"agency":"", "title":"", "prodNo1":"", "prodNo2":"", "tour":{}, "href":""}
    
    headers = {
    'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Accept-Encoding':'gzip, deflate, sdch',
    'Accept-Language':'zh-TW,zh;q=0.8,en-US;q=0.6,en;q=0.4',
    'Cache-Control':'max-age=0',
    'Connection':'keep-alive',
    'Cookie':'JSESSIONID=036013613DF1E94DEDAE39EF6C952CC9; __BWfp=c1463396708123x7c907c0ce; _ST.REC.UID=aae3edf9-fc65-4c43-9cfe-35aaa93b77d7; __lastv=GFG0000008930; _dc_gtm_UA-27813386-1=1; __asc=55ce0033154e5cd7bb0641af45a; __auc=4157e5c2154b93c1502ced98a9a; _ga=GA1.3.464529237.1463396604; __eruid=1e55da88-33f9-36c9-de3e-7423951404fe',
    'Host':'www.settour.com.tw',
    'Referer':'https://www.settour.com.tw/',
    'Upgrade-Insecure-Requests':'1',
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
    }
    
    res = requests.get(url, headers = headers)  #要透過headers才抓的到資料
    soup =  bs(res.text, 'lxml')
    num += 1   #計算總共行程數
    #title  title裡會有不相干的字，用slipt剔除
    dic["title"] = soup.select('div.pd-content h1')[0].text.strip().split(u"團")[0]
    
    #agency
    dic["agency"] = u"東南旅遊"
    
    #prodNo   將網址切顆只剩下prodNo, grupNo
    rear = url.split('&erpOrdInfo=')  #
    rearstring = rear[0].split('prodNo=')
    prodNo, grupNo =  rearstring[1].split('&grupNo=')  
    dic["prodNo1"] = grupNo
    dic["prodNo2"] = prodNo
    
    #href
    dic["href"] = url
      
    for i in range( 0, len(soup.select('div.schedule h3'))):      
    #tour  每天的行程不同，用day1..day+n來區分
        dic["tour"]['Day'+str(i+1)] = soup.select('div.schedule h3')[i].text.split()[1] 

    tourlist.append(dic)  #將dic加入tourlist裡
```
